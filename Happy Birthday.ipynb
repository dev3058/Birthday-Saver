# Happy Birthday

```python
import json
def decode():
    try:
        file=open('data.json','r')
        data=json.load(file)
        file.close()
        return data
    except FileNotFoundError:
        file=open('data.json','w+')
        empty={}
        json.dump(empty,file)
        file.close()
        return empty
        
def encode(obj):
    file=open('data.json','w')
    json.dump(obj,file)
    file.close()

print("\t\t Happy Birthday List")
print("1. Show the Birthday")
print("2. Add the Birthday")
print("3. Update the Birthday")
print("4. Delete the Birthday")
print("5. See the Birthday List")
print("6. Exit")
print()
print()
lit=decode()
print(type(lit))
while True:
    cond=int(input("Enter the Condition: "))
    if cond==1:
        if lit=={}:
            print("List Is Empty")
        else:
            nw=input("Enter the Name: ")
            bw=lit.get(nw,"Name Not Found")
            print(bw)
    elif cond==2:
        new=input("Enter the Name Of Friend : ")
        bith=input("Enter the Birthday: ")
        lit[new]=bith
        print("Birthday Added Successfully")
    elif cond==3:
        nuw=input("Enter the Name Of Friend : ")
        if nuw in lit:
            buth=input("Enter New Birthday: ")
            lit[nuw]=buth
            print("Birthday Updated Successfully")
        else:
            print("Name Not Found")
    elif cond==4:
        ndw=input("Enter the Name Of Friend : ")
        if ndw in lit:
            cn=input("Do You really Want to Delete the Data Y/N ?")
            if cn=="y" or cn=="Y":
                del lit[ndw]
                print("Birthday Deleted Successfully")
            else:
                print("Birthday is Not Deleted Yet")
        else:
            print("Name Not Found")
    elif cond==5:
        if lit=={}:
            print("List Is Empty")
        else:
            print(lit)
    
    elif cond==6:
        print("Thanks For Visit ")
        encode(lit)
        break
    
    else:
        print("Wrong Condition has Chosen")
```
